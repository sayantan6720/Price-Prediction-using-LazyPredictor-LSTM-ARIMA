# **LSTM**

Next, LSTM is used. For the "x-values", 60 days of data has been provided using which the model predicts the price for the 61st day, the "y-value". The data has been scaled using MinMaxScaler to make it easier for the model to process the data.

The model contains 4 layers including the input layer and 1 output layer. Dropout rate has been set to 0.2. The loss function used to evaluate the model is mean sqaured error. 

```python
df1=df.iloc[0:276, 0:2]
df1
```

```python
df1.rename(columns = {"Domestic Market (Contract) Blow Molding, Low":"low price"}, inplace=True)
df1
```

```python
df1.plot()
```

```python
len_train=int(0.75*len(df1["low price"]))
train=df1.iloc[0:len_train,1:2].values
test=df1.iloc[len_train:,1:2].values

print(len(train))
print(len(test))
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
train_scaled = sc.fit_transform(train)
print(train_scaled)
type(train_scaled)
```

```python
xtrain=[]
ytrain=[]

for i in range(60, len(train)):
    xtrain.append(train_scaled[i-60: i, 0])
    ytrain.append(train_scaled[i, 0])
xtrain, ytrain = np.array(xtrain), np.array(ytrain)
xtrain
```

```python
import tensorflow as tf
from tensorflow import keras
import keras
from keras.layers import Dense
from keras.utils import to_categorical
from keras.layers import LSTM
import seaborn as sns
from keras.layers import Dropout
```

```python
model=keras.Sequential()
```

```python
model.add(LSTM(units = 64, return_sequences = True, input_shape = (xtrain.shape[1], 1)))
model.add(Dropout(rate = 0.2))
```

```python
##2nd lstm layer
model.add(LSTM(units = 64, return_sequences = True))
model.add(Dropout(rate = 0.2))
##3rd lstm layer
model.add(LSTM(units = 64, return_sequences = True))
model.add(Dropout(rate = 0.2))
##4th lstm layer
model.add(LSTM(units = 64, return_sequences = False))
model.add(Dropout(rate = 0.2))
```

```python
#Output layer
model.add(Dense(units = 1))
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
)
```

```python
model.fit(
    xtrain,
    ytrain, 
    batch_size=32,
    epochs=100
    )
```

```python
inp=df1.iloc[len(train)-60:,1:2].values
print(len(inp))
print(inp)
```

```python
inp=sc.transform(inp)
inp
```

```python
xtest=[]
for i in range(60, len(inp)):
  xtest.append(inp[i-60: i, 0])

xtest = np.array(xtest)
```

```python
pred_scaled=model.predict(xtest)
```

```python
pred=sc.inverse_transform(pred_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(test, pred)
r2 = r2_score(test,pred)

print("MSE: ", mse)
print("R2: ", r2)
```

```python
plt.plot(test, color = 'red', label = 'Original')
plt.plot(pred, color = 'blue', label = 'Predicted')
plt.title('Low Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

# **Univariate ARIMA**

Next ARIMA has been used for forecasting the price. We use a ARIMA model with p,d,q values as 2,1,1 which we found to be the optimal parameters using Akaike Information Criterion (AIC). 

```python
df1.plot()
```

```python
df_arima = df1.iloc[0:,1:2]
df_arima
```

```python
df_arima = np.log(df_arima)
df_arima.plot()
```

```python
train_arima=df_arima.iloc[0:len_train,0:]
test_arima=df_arima.iloc[len_train:,0:]

print(len(train_arima))
print(len(test_arima))
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

acf_o=plot_acf(train_arima)
pacf_o=plot_pacf(train_arima)
```

```python
train_arima_diff=train_arima.diff().dropna()
train_arima_diff.plot()
```

```python
acf_o=plot_acf(train_arima_diff)

pacf_o=plot_pacf(train_arima_diff)
```

```python
from statsmodels.tsa.arima.model import ARIMA

aic_min=np.inf

for p_value in range(0,4):
  for d_value in range(0,3):
    for q_value in range(0,4):
      model = ARIMA(train_arima, order=(p_value, d_value, q_value))
      results=model.fit()
      aic=results.aic
    
      if aic < aic_min:
        aic_min = aic
        optimal=(p_value, d_value, q_value)
    
          
print(f"Optimal order: {optimal}, AIC: {aic_min}")      
```

```python
model_final = ARIMA(train_arima, order=(2,1,1))
model_fit=model_final.fit()
```

```python
residuals = model_fit.resid[1:]
fig, ax = plt.subplots(1,2)
residuals.plot(title='Residuals', ax=ax[0])
residuals.plot(title='Density', kind='kde', ax=ax[1])
plt.show()
```

```python
plot_acf(residuals)
plot_pacf(residuals)
```

```python
forecast=model_fit.forecast(len(test_arima))
forecast1=model_fit.get_forecast(len(test_arima))
conf_int=forecast1.conf_int(alpha=0.05)
```

```python
plt.plot(test_arima, color = 'red', label = 'Original')
plt.plot(forecast, color = 'blue', label = 'Predicted')
plt.plot(conf_int, label="confidence interval")
plt.title('Low Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
mse_arima=mean_squared_error(test_arima, forecast)
r2_arima=r2_score(test_arima, forecast)

print("MSE: ", mse_arima,
      "R2: ", r2_arima)
```
